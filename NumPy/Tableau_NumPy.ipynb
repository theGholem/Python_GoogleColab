```python
# '.linspace(start, stop, num)' : fonction de la biblio de NumPy qui permet d'epacer ou diviser unformement l'intervalle [start, stop]

np.linspace(0, 10, 2) # on veut obtenir 2 nombres dans [0, 10] uniformenent

# Naturellement la sortie est de type float
```

```python
np.linspace(0, 10, 10) # on veut obtenir 10 nombres dans l'intervalle [0, 10] uniformement
```

```python
# Supposons que on veut la meme sortie du dessu, mais avec des resultats fractionnel, cad pas avec autant de nombres apres la virgule
# Dans ce cas , il faudra ajouter 1 à notre nombre 'stop'

np.linspace(0, 10, 11)
```

```python
# Pour mieux visualiser

print(' Sans forme fractionnelle :\n')
np.linspace(0, 10, 20)
print(' Avec forme fractionnelle :\n')
np.linspace(0, 10, 21)
```

```python
# on peut generer aleatoirement des nombre flottant biensur dans des matrices selon la loi uniforme
# cad nos valeur sont uniformement repartie

np.random.rand(3) # genere 3 nombre flottant uniformement reparti
```

```python
# random.rand(line, columns)

np.random.rand(3, 4)
```

```python
# distribution normal standard
# moyenne=0 (centré sur 0), variance=1
# genere des nombres negatifs compris entre -1 et 1, et biensur des nombres positifs à partir de 0

np.random.randn(4, 4)
```

```python
# generer aleatoirement un nombre a partir de l'intervalle [a, b], avec a inclu et b pas inclu

np.random.randint(0, 100) # 0 peut etre selectionné, mais jamais 100
```

```python
# on veut maintenant preciser le nombre n d'element que on veut recuperer : .randint(a, b, n)

np.random.randint(0, 101, 20) # On recupere 20 elements avec 100 inclut et 101 exclut
```

```python
# Mais on peut aussi faire afficher nos nombres dans un tableau matricielle (x_line, y_colone): .randint(a, b, (x,y) )

np.random.randint(0, 100, (5,5)) # 5 line, 5 columns
```

```python
# pour reproduire les memes nombres aleatoires
# .random.seed(id_Number) : c est comme un id , pour tel id, la machine genere des nombres identique pour tous le monde

np.random.seed(42)
np.random.rand(4)

# faut toujours les executer ensembles pour que le 'seed()' puisse fixer la valeur
```

```python
# si on execute en solo, ca ne fonctionne pas, donc faut toujours preceder cela d'un seed()
# A ce moment ici, il faut toujours executer encore une fois le code ci dessus avec le seed() afin d'avoir des valeurs aleatoire
# identique. cad on pourra garder la fonctionnalité du seed() qui genere toujours les memes variables pour tous le monde

np.random.rand(4)
```
