```python
import numpy as np
```

```python
# intialisation de notre tableau

arr=np.arange(11)
```

```python
# affichage

arr
```

```python
# operation de base +

arr+5 # ajoute 5 à chaque element de notre tableau
```

```python
# operation de base -

arr-5
```

```python
# operation de base *

arr*5 # 5*arr
```

```python
# operation de base /

arr/5 # on obtient biensur un float
```

```python
# on peut egalement faire les operations sur les tableaux directement

arr+arr # arr-arr, arr*arr, arr/arr : a ce momment les operations se deroulent entre chaque element
```

```python
arr-arr # vaut un tableau 0 naturellement
```

```python
# la division par 0 etant impossible

# arr/0 ou arr/arr car on a 0 dans notre tableau aumoin 1 fois

arr/arr # NumPy fournit un resultat et un avertissement et calcule les autre sauf ou il y a le probleme(en mettant 'nan')
```

```python
# si par contre c'est directement un entier qui divise 0, alors il retourne egalement une erreur partielle et
# ajoute un 'inf' (infini) a la place du probleme

1/arr
```

```python
# on peut aussi utiliser les fonction mathematique pures

# pour la racine carrée ( sqrt)

np.sqrt(arr)
```

```python
# c'est pareil pour le sinus ou le cosinus

np.sin(arr)
```

```python
# pour le logarithme, on aura une erreure partielle car dans notre tableau on a 0 et on sait que log(0) n'est pas defini mais tend vers -inf

np.log(arr)
```
